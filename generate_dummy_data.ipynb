```python
# Initialize Faker
fake = Faker('en_US')  # Use Indonesian locale
```

```python
# Base datetime
base_date = datetime(2025, 10, 20, 10, 13, 37, 758306)
```

```python
fungsi_enum = ["STAFF", "GURU_HURIA", "BIBELVROUW", "CALON_DIAGONES", "DIAGONES", "CALON_SINTUA","SINTUA","CALON_PENDETA","PENDETA"]
status_enum = ["PENDIDIKAN", "ACTIVE", "PENSIUN"]
```

```python
# Generate data
dummy_data = []
for i in range(113, 150): # must be + 1 from id needed


    # year = random.randint(2025, 2025) # bisa dibuat range tahun 1998 - 2025 bebas aja (1998, 2025)
    # month = random.randint(10, 12)
    # day = random.randint(1, 28)
    # hour = random.randint(0, 23)
    # minute = random.randint(0, 59)
    # second = random.randint(0, 59)

    birth_year = random.randint(1960, 2024)
    # times = datetime(
    #     year, month, day, hour, minute, second
    # )
    birth_date = datetime(birth_year, random.randint(10, 10), random.randint(23, 23), random.randint(0, 23), random.randint(0, 59), random.randint(0, 59))

    age = base_date.year - birth_date.year - ((base_date.month, base_date.day) < (birth_date.month, birth_date.day))

    entry = {
        "id": i,
        # "activity_date": times.strftime("%Y-%m-%d 00:00:00"),
        # "activity_time": times.strftime("%H:%M:%S"),
        # "activity_title": fake.sentence(),
        # "description": fake.paragraph(),
        # "location": fake.city(),
        # "pic": fake.name(),
        # "time_hour": times.strftime("%H Jam %M Menit"),
        "name": fake.first_name(),
        "birth_date": birth_date.strftime("%Y-%m-%d %H:%M:%S"),
        "phone_number": f"+62{random.randint(81100000000, 89999999999)}",
        "address": fake.address().replace("\n", ", "),
        "age": age,
        "fungsi": random.choice(fungsi_enum),    # random enum value
        "status": random.choice(status_enum),    # random enum value
        "created_at": base_date.strftime("%Y-%m-%d %H:%M:%S.%f"),
        "updated_at": base_date.strftime("%Y-%m-%d %H:%M:%S.%f")
    }
    dummy_data.append(entry)
```

```python
# Print as SQL insert statements
for entry in dummy_data:
    sql = f"""
INSERT INTO master_board ("id", "name", "birth_date", "phone_number", "address", "age", "fungsi", "status", "created_at", "updated_at")
VALUES (
    {entry['id']},
    '{entry['name']}',
    '{entry['birth_date']}',
    '{entry['phone_number']}',
    '{entry['address']}',
    '{entry['age']}',
    '{entry['fungsi']}',
    '{entry['status']}',
    '{entry['created_at']}',
    '{entry['updated_at']}'
);
""".strip()
    print(sql)
```
